```python
"""IMPORTING MODULES AND DATASET"""
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
import networkx as nx
import scipy
import numpy as np
```

```python
def read_integers(path):
    content_lst=[]
    with open(path,'r') as file:
        content=file.read()
        for i in content.split('\n'):
            content_lst+=i.split(', ')
        return content_lst

names = read_integers('patient.txt')
names.pop()
print(names)
patient_number = len(names)
print(patient_number)
```

```python
"""THIS PART IS FOR LOADING SEMA'S DATASET"""
ts = np.loadtxt('HCP dataset-56 fMRI/TS_HCP/subj_104416_ts', delimiter = ',')
sig = []
for i in range(patient_number):
    sig.append(np.loadtxt('HCP dataset-56 fMRI/TS_HCP/subj_' + str(names[i]) +'_ts', delimiter = ','))
```

```python
"""THIS PART IS FOR LOADING YOUR OWN DATASET"""
G = []
for i in range(patient_number):
    G.append(np.loadtxt(open('/home/duc/connectorhub/Connectome (new)-20240131T201631Z-001/Connectome (new)/subj_' + str(names[i]) + '_mmp1.csv','rb'), delimiter=",")[0:360,0:360])
print(G)
```

```python
import scipy
info = scipy.io.loadmat("eeg_signal.mat")
sig = info["f"]
G = info["A"]
```

```python
G_nx = []
for i in range(patient_number):
    G_nx.append(nx.from_numpy_array(G[i])) #creating graph from the patients
print(len(G_nx))
print(G_nx)
```

```python
scipy.io.savemat(file_name = "adjacency.mat", mdict = {'A_nx': G})
```

```python
#calculating the community partition of every patient
import numpy as np 
comm = []
for k in range(patient_number):
    comm_object = nx.community.louvain_communities(G_nx[k], seed = 1)
    for i in range(len(comm_object)):
        comm_object[i] = np.array(list(comm_object[i]))
    print(np.array(comm_object, dtype = 'object'))
    comm.append(np.array(comm_object, dtype = 'object'))
comm = np.asarray(comm, dtype = 'object')
print(comm)
scipy.io.savemat(file_name = "comm.mat", mdict = {'comm0': comm})
```

```python
comm_matrix = []
for k in range(patient_number):
    subgraph_matrix = []
    for i in range(len(comm[k])):
        subgraph_matrix.append(nx.adjacency_matrix(G_nx[k].subgraph(comm[k][i])))
    comm_matrix.append(subgraph_matrix)
comm_matrix_object = np.asarray(comm_matrix, dtype="object")           #This fixes the bug where you can't save the matrix because it isn't a homogenous list
print(comm_matrix_object)
scipy.io.savemat(file_name = "comm_matrix.mat", mdict = {'matrices': comm_matrix_object})
```
